```python
import pandas as pd
import random

data = [[random.random() for i in range(10)] for j in range (1,8)]
df   = pd.DataFrame (data).round(3)
df.columns = ['Column '+str(i) for i in range(1,11)]
lista = [df.columns[:,].values.astype(str).tolist()] + df.values.tolist()
df
```

```python
from reportlab.lib import colors
from reportlab.lib.pagesizes import letter,A4
from reportlab.platypus import Table, TableStyle, Paragraph, ListFlowable, ListItem
from reportlab.pdfgen.canvas import Canvas
from reportlab.lib.styles import getSampleStyleSheet,ParagraphStyle
from reportlab.lib.enums import TA_LEFT,TA_CENTER,TA_JUSTIFY

from reportlab.lib.units import mm

from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont
from reportlab.lib.fonts import addMapping


#################### code for register font ####################
font = 'OpenSans'
pdfmetrics.registerFont(TTFont('OpenSans-Regular', 'OpenSans-Regular.ttf'))
pdfmetrics.registerFont(TTFont('OpenSans-Italic', 'OpenSans-Italic.ttf'))
pdfmetrics.registerFont(TTFont('OpenSans-Bold', 'OpenSans-Bold.ttf'))
pdfmetrics.registerFont(TTFont('OpenSans-BoldItalic', 'OpenSans-BoldItalic.ttf'))
addMapping(font, 0, 0, 'OpenSans-Regular')
addMapping(font, 0, 1, 'OpenSans-Italic')
addMapping(font, 1, 0, 'OpenSans-Bold')
addMapping(font, 1, 1, 'OpenSans-BoldItalic')

font = 'OpenSansBold'
pdfmetrics.registerFont(TTFont('OpenSansBold-Bold', 'OpenSans-ExtraBold.ttf'))
pdfmetrics.registerFont(TTFont('OpenSansBold-BoldItalic', 'OpenSans-ExtraBoldItalic.ttf'))
pdfmetrics.registerFont(TTFont('OpenSansBold-Regular', 'OpenSans-SemiBold.ttf'))
pdfmetrics.registerFont(TTFont('OpenSansBold-Italic', 'OpenSans-SemiBoldItalic.ttf'))
addMapping(font, 0, 0, 'OpenSansBold-Regular')
addMapping(font, 0, 1, 'OpenSansBold-Italic')
addMapping(font, 1, 0, 'OpenSansBold-Bold')
addMapping(font, 1, 1, 'OpenSansBold-BoldItalic')


#################### code for stylesheet ####################
styles = getSampleStyleSheet()
MyStyleSheets = {}


MyStyleSheets['Normal'] = ParagraphStyle('Normal')
MyStyleSheets['Normal'].spaceAfter = 4
MyStyleSheets['Normal'].fontName = 'OpenSans-Regular'
MyStyleSheets['Normal'].fontSize= 8
MyStyleSheets['Normal'].textColor =  '#434343'
MyStyleSheets['Normal'].alignment = TA_JUSTIFY


MyStyleSheets['title'] = ParagraphStyle('Title')
MyStyleSheets['title'].fontName = 'OpenSansBold-Bold'
MyStyleSheets['title'].alignment = TA_CENTER
MyStyleSheets['title'].spaceAfter = 4
MyStyleSheets['title'].fontSize = 24
MyStyleSheets['title'].textColor =  '#434343'


MyStyleSheets['Header1'] = ParagraphStyle('Heading1')
MyStyleSheets['Header1'].spaceAfter = 4
MyStyleSheets['Header1'].fontName = 'OpenSansBold-Bold'
MyStyleSheets['Header1'].fontSize=18
MyStyleSheets['Header1'].textColor =  '#434343'

#################### code for inistiation and adding header ####################
canv = Canvas("doc.pdf", pagesize=A4)
title_line_1 = Paragraph("My HitMiss Report for Richard Xia", MyStyleSheets['title'])
header_line_1 = Paragraph("Table 1", MyStyleSheets['Header1'])
header_line_2 = Paragraph("Table 2", MyStyleSheets['Header1'])

intro_para = Paragraph("Lorem ipsum dolor sit amet, consectetur adipiscing elit, sed do eiusmod tempor incididunt ut labore et dolore magna aliqua. Ut enim ad minim veniam, quis nostrud exercitation ullamco laboris nisi ut aliquip ex ea commodo consequat. Duis aute irure dolor in reprehenderit in voluptate velit esse cillum dolore eu fugiat nulla pariatur. Excepteur sint occaecat cupidatat non proident, sunt in culpa qui officia deserunt mollit anim id est laborum.", MyStyleSheets['Normal'])
sample_string="Lorem ipsum dolor sit amet, consectetur adipiscing elit, sed do eiusmod tempor incididunt ut labore et dolore magna aliqua. Ut enim ad minim veniam, quis nostrud exercitation ullamco laboris nisi ut aliquip ex ea commodo consequat."
bulletPoints = [sample_string,'point2','point3']

header_t = Table([['']],(5),(25))
header_t.setStyle(TableStyle([
                        ('FONT', (0,0), (-1,0), 'OpenSansBold-Regular'),
                        ('FONTSIZE', (0,0), (-1,-1), 42),
                        ('BACKGROUND', (0, 0), (-1, -1), '#2d5254')
                      ]))


#################### code for table ####################
t = Table(lista)
t.setStyle(TableStyle([ ('ALIGN', (0,0), (-1,-1), 'CENTER'), 
                        ('LINEAFTER', (0,0), (-2,0), 0.25, '#a3a3a3', None, None, 1, 1, 0.5),
                        ('LINEBELOW', (0,0), (-1,0), 1, '#242424', None, None, 1, 1, 2),
                        ('FONT', (0,0), (-1,0), 'OpenSansBold-Regular'),
                       ('FONT', (0,1), (-1,-1), 'OpenSans-Regular'),
                        ('FONTSIZE', (0,0), (-1,-1), 8)
                      ]))


'''
# line commands are like
# op, start, stop, weight, colour, cap, dashes, join, linecount, linespacing

LINEABOVE
LINEBELOW
LINEBEFORE
LINEAFTER
'''



data_len = len(lista)

for each in range(data_len):
    if each % 2 == 1:
        bg_color = colors.white
    else:
        bg_color = '#e9f2f5'
    t.setStyle(TableStyle([('BACKGROUND', (0, each), (-1, each), bg_color)]))

t.setStyle(TableStyle([('BACKGROUND', (0, 0), (-1, 0), '#ffffff')]))    
    
    
####################  code for generating pdf     ####################
aW,aH=A4
marginLeft=15
w, h = header_t.wrap(aW, aH)


aH = aH - 40
w, h = title_line_1.wrap(aW, aH)
h+=20
title_line_1.drawOn(canv, 0, aH)



aH = aH - h
w, h = intro_para.wrap(aW-80, aH)
h+=20
intro_para.drawOn(canv,marginLeft+20, aH-h)


for point in bulletPoints:
    aH = aH - h
    bulletSign = Paragraph('\u2022', MyStyleSheets['Normal'])
    bulletPoint = Paragraph(point, MyStyleSheets['Normal'])
    _ = bulletSign.wrap(aW, aH)
    w, h = bulletPoint.wrap(aW-90, aH)
    h+=2
    bulletSign.drawOn(canv,marginLeft+20, aH-14)
    bulletPoint.drawOn(canv,marginLeft+30, aH-h)

aH = aH - h
w, h = intro_para.wrap(aW-80, aH)
h+=20
intro_para.drawOn(canv,marginLeft+20, aH-h)

aH = aH - h
w, h = header_line_1.wrap(aW, aH)
h+=20
header_t.drawOn(canv, marginLeft, aH-h-12)
header_line_1.drawOn(canv, marginLeft+20, aH-h)

aH = aH - h
w, h = t.wrap(aW, aH)
h+=20
t.drawOn(canv,marginLeft+20, aH-h)


aH = aH - h
w, h = header_line_2.wrap(aW, aH)
h+=20
header_t.drawOn(canv, marginLeft, aH-h-12)
header_line_2.drawOn(canv, marginLeft+20, aH-h)


aH = aH - h
w, h = t.wrap(aW, aH)
h+=20
t.drawOn(canv, marginLeft+20, aH-h)

canv.save()
```
